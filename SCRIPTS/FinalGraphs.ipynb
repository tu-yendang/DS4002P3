# Final Graphs:
This takes the final_training_CNN.csv file and graphs the loss and accuracy as time (measured by epochs) continue.

```python
# import statements
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file
data = pd.read_csv("final_training_CNN.csv" )
```

```python
# graph for loss
plt.figure(figsize=(10, 5))
plt.plot(data['Epoch'], data['Train Loss'], label='Train Loss', color='blue')
plt.plot(data['Epoch'], data['Validation Loss'], label='Validation Loss', color='red')
plt.title('Train Loss vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# graph for accuracy
plt.figure(figsize=(10, 5))
plt.plot(data['Epoch'], data['Train Accuracy'], label='Train Accuracy', color='blue')
plt.plot(data['Epoch'], data['Validation Accuracy'], label='Validation Accuracy', color='red')
plt.title('Train Accuracy vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
