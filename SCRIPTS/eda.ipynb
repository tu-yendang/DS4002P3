# EDA:
This document is for our exploratory data analysis, where we focus on the number of books each genre has for our training data set. It can be seen that almost all of the 13 genres had book data points, but young adult fantasy was missing one value, and humor missing 20 values (due to the 2024 GoodRead's Awards removing Humor as a book genre).

```python
# import statements
import csv
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# open file and count each genre
file_path = '/content/train_book_covers.csv'
genre_counter = Counter()

with open(file_path,'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        genre = row['Genre']
        genre_counter[genre] += 1

# Plot the genre counts
genres = list(genre_counter.keys())
counts = list(genre_counter.values())

plt.figure(figsize=(10, 6))
plt.bar(genres, counts, color='skyblue')
plt.xlabel('Genres')
plt.ylabel('Number of Books')
plt.title('Number of Books by Genre')
plt.xticks(rotation=90)
plt.tight_layout()

# save plot
plt.savefig('eda_book_count_by_genre.png')

plt.show()
```
